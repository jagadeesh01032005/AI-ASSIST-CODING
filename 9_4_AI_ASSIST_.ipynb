```python
def calculate_average(numbers):
    # Initialize total sum to zero for accumulation
    total_sum = 0
    # Initialize count of numbers to zero
    count = 0
    # Loop through each number in the input list
    for num in numbers:
        # Add the current number to the total sum
        total_sum += num
        # Increment the count of numbers processed
        count += 1
    # Check if count is zero to avoid division by zero error
    if count == 0:
        # If no numbers, return 0 as average
        return 0
    else:
        # Calculate the average by dividing total sum by count
        average = total_sum / count
        # Return the calculated average
        return average

def reverse_string(s):
    # Initialize an empty string to store the reversed string
    reversed_s = ""
    # Loop through the input string from the end to the beginning
    # len(s) - 1 is the index of the last character
    # -1 is the stop index (before the first character)
    # -1 is the step, meaning decrement by 1 each time
    for i in range(len(s) - 1, -1, -1):
        # Append the current character to the reversed string
        reversed_s += s[i]
    # Return the newly formed reversed string
    return reversed_s
```

```python
def calculate_average(numbers: list) -> float:
    """Calculates the average of a list of numbers.

    This function iterates through a list of numbers, sums them up, and then
    divides by the count to find the average. It handles an empty list by
    returning 0 to prevent a division by zero error.

    Args:
        numbers (list): A list of numerical values.

    Returns:
        float: The calculated average of the numbers, or 0 if the list is empty.

    Examples:
        >>> calculate_average([1, 2, 3, 4, 5])
        3.0
        >>> calculate_average([])
        0
    """
    total_sum = 0
    count = 0
    for num in numbers:
        total_sum += num
        count += 1
    if count == 0:
        return 0
    else:
        average = total_sum / count
        return average

def reverse_string(s: str) -> str:
    """Reverses a given string.

    This function iterates through the input string from the last character
    to the first, appending each character to a new string to form the
    reversed version.

    Args:
        s (str): The input string to be reversed.

    Returns:
        str: The reversed string.

    Examples:
        >>> reverse_string("hello")
        "olleh"
        >>> reverse_string("Python")
        "nohtyP"
    """
    reversed_s = ""
    for i in range(len(s) - 1, -1, -1):
        reversed_s += s[i]
    return reversed_s
```

```python
import ast

def read_and_parse_python_file(file_path: str):
    """Reads a Python file and parses its content into an Abstract Syntax Tree (AST).\

    Args:
        file_path (str): The path to the Python file.

    Returns:
        tuple: A tuple containing the AST object and the source code string.
               Returns (None, None) if FileNotFoundError occurs.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            source_code = file.read()
        ast_tree = ast.parse(source_code)
        print(f"Successfully read and parsed file: {file_path}")
        return ast_tree, source_code
    except FileNotFoundError:
        print(f"Error: The file at '{file_path}' was not found.")
        return None, None
    except Exception as e:
        print(f"An unexpected error occurred while processing '{file_path}': {e}")
        return None, None

print("Defined function 'read_and_parse_python_file' for reading and parsing Python files.")
```

```python
import ast

def find_definitions(ast_tree):
    """Traverses an AST to find function and class definitions.

    Args:
        ast_tree: The Abstract Syntax Tree object.

    Returns:
        tuple: Two lists, one containing ast.FunctionDef nodes and one containing ast.ClassDef nodes.
    """
    functions = []
    classes = []
    for node in ast.walk(ast_tree):
        if isinstance(node, ast.FunctionDef):
            functions.append(node)
        elif isinstance(node, ast.ClassDef):
            classes.append(node)
    print("Defined function 'find_definitions' for traversing the AST.")
    return functions, classes

print("Defined function 'find_definitions' for traversing the AST.")
```

```python
import ast

def _generate_function_docstring(func_node: ast.FunctionDef) -> str:
    """Generates a Google-style docstring for a function node."""
    func_name = func_node.name
    docstring_parts = [f""""""A brief description of the `{func_name}` function.

    """"]

    # Parameters
    if func_node.args.args or func_node.args.kwonlyargs or func_node.args.posonlyargs:
        docstring_parts.append("Args:")
        for arg in func_node.args.posonlyargs + func_node.args.args + func_node.args.kwonlyargs:
            arg_name = arg.arg
            arg_type = 'Any' # Default type
            if arg.annotation:
                # Attempt to get the type name from the annotation node
                if isinstance(arg.annotation, ast.Name):
                    arg_type = arg.annotation.id
                elif isinstance(arg.annotation, ast.Constant):
                    arg_type = str(arg.annotation.value)
                elif isinstance(arg.annotation, ast.Subscript):
                    # Handles types like List[int], Dict[str, Any], etc.
                    if isinstance(arg.annotation.value, ast.Name):
                        type_name = arg.annotation.value.id
                        if isinstance(arg.annotation.slice, ast.Index) and isinstance(arg.annotation.slice.value, ast.Name):
                            inner_type = arg.annotation.slice.value.id
                            arg_type = f"{type_name}[{inner_type}]"
                        elif isinstance(arg.annotation.slice, ast.Tuple):
                            inner_types = []
                            for elt in arg.annotation.slice.elts:
                                if isinstance(elt, ast.Name):
                                    inner_types.append(elt.id)
                                elif isinstance(elt, ast.Constant):
                                    inner_types.append(str(elt.value))
                            arg_type = f"{type_name}[{', '.join(inner_types)}]"
                        else:
                            arg_type = 'Any'
                    else:
                        arg_type = 'Any'

            docstring_parts.append(f"        {arg_name} ({arg_type}): Description of `{arg_name}`.")

    # Return type
    return_type = 'None' # Default return type for functions without explicit return
    if func_node.returns:
        if isinstance(func_node.returns, ast.Name):
            return_type = func_node.returns.id
        elif isinstance(func_node.returns, ast.Constant):
            return_type = str(func_node.returns.value)
        elif isinstance(func_node.returns, ast.Subscript):
            if isinstance(func_node.returns.value, ast.Name):
                type_name = func_node.returns.value.id
                if isinstance(func_node.returns.slice, ast.Index) and isinstance(func_node.returns.slice.value, ast.Name):
                    inner_type = func_node.returns.slice.value.id
                    return_type = f"{type_name}[{inner_type}]"
                elif isinstance(func_node.returns.slice, ast.Tuple):
                    inner_types = []
                    for elt in func_node.returns.slice.elts:
                        if isinstance(elt, ast.Name):
                            inner_types.append(elt.id)
                        elif isinstance(elt, ast.Constant):
                            inner_types.append(str(elt.value))
                    return_type = f"{type_name}[{', '.join(inner_types)}]"
                else:
                    return_type = 'Any'
            else:
                return_type = 'Any'

    if func_node.returns:
         docstring_parts.append(f"\n    Returns:
        {return_type}: Description of the return value.")

    docstring_parts.append(""""""") # Closing triple quotes
    return '\n'.join(docstring_parts)

def _generate_class_docstring(class_node: ast.ClassDef) -> str:
    """Generates a Google-style docstring for a class node."""
    class_name = class_node.name
    return f""""""A class representing a {class_name}.

    This class is designed to [briefly describe its purpose].
    """"""" # Ensure proper escaping for triple quotes

print("Defined helper functions '_generate_function_docstring' and '_generate_class_docstring'.")
```

```python
import ast

def _get_type_string(annotation_node):
    """Helper to extract a readable type string from an AST annotation node."""
    if annotation_node is None:
        return 'Any' # Default if no annotation is present
    if isinstance(annotation_node, ast.Name):
        return annotation_node.id
    if isinstance(annotation_node, (ast.Constant, ast.Str)): # ast.Str for older Python versions
        return str(annotation_node.value)
    if isinstance(annotation_node, ast.Subscript):
        type_name = _get_type_string(annotation_node.value)
        if isinstance(annotation_node.slice, ast.Index):
            inner_type = _get_type_string(annotation_node.slice.value)
            return f"{type_name}[{inner_type}]"
        elif isinstance(annotation_node.slice, ast.Tuple):
            inner_types = [_get_type_string(elt) for elt in annotation_node.slice.elts]
            return f"{type_name}[{', '.join(inner_types)}"]"
    return 'Any' # Fallback for complex or unhandled types

def _generate_function_docstring(func_node: ast.FunctionDef) -> str:
    """Generates a Google-style docstring for a function node."""
    func_name = func_node.name
    content = []

    content.append(f"A brief description of the `{func_name}` function.")

    # Parameters
    params_exist = bool(func_node.args.posonlyargs or func_node.args.args or func_node.args.kwonlyargs)
    if params_exist:
        content.append("") # Blank line before Args
        content.append("Args:")
        for arg in func_node.args.posonlyargs + func_node.args.args + func_node.args.kwonlyargs:
            arg_name = arg.arg
            arg_type = _get_type_string(arg.annotation)
            content.append(f"    {arg_name} ({arg_type}): Description of `{arg_name}`.")

    # Return type
    return_type = _get_type_string(func_node.returns)
    content.append("") # Blank line before Returns
    content.append("Returns:")
    content.append(f"    {return_type}: Description of the return value.")

    # Examples
    content.append("") # Blank line before Examples
    content.append("Examples:")
    content.append(f"    >>> # Example usage for `{func_name}`")

    # Join lines and wrap in triple quotes
    return '"""\n' + '\n'.join(content) + '\n"""'

def _generate_class_docstring(class_node: ast.ClassDef) -> str:
    """Generates a Google-style docstring for a class node."""
    class_name = class_node.name
    docstring_content = f"A class representing a {class_name}.\n\n    This class is designed to [briefly describe its purpose]."
    return f'"""\n{docstring_content}\n"""'

print("Defined helper functions '_generate_function_docstring' and '_generate_class_docstring'.")
```

```python
import ast

class DocstringInjector(ast.NodeTransformer):
    """Inserts Google-style docstrings into function and class definitions if they don't already have one."""

    def visit_FunctionDef(self, node: ast.FunctionDef) -> ast.FunctionDef:
        """Visits a FunctionDef node and inserts a docstring if missing."""
        # Check if the function already has a docstring
        if (node.body and
                isinstance(node.body[0], ast.Expr) and
                isinstance(node.body[0].value, (ast.Constant, ast.Str))):
            # Docstring already exists, do nothing
            return node

        # Generate a new docstring
        new_docstring_content = _generate_function_docstring(node)

        # Create an AST node for the docstring
        # ast.Constant is preferred for Python 3.8+ for string literals
        # ast.Str is for older Python versions
        docstring_expr = ast.Expr(value=ast.Constant(value=new_docstring_content))

        # Copy line and column information from the function node for proper placement
        ast.copy_location(docstring_expr, node)
        # Ensure the docstring is indented correctly relative to the function body
        # This often involves adjusting col_offset, but for insertion, it usually follows
        # the function's own indentation. The `ast.unparse` will handle final formatting.

        # Insert the docstring at the beginning of the function body
        node.body.insert(0, docstring_expr)

        # Ensure the function definition itself maintains its own location info
        ast.fix_missing_locations(node)

        return node

    def visit_ClassDef(self, node: ast.ClassDef) -> ast.ClassDef:
        """Visits a ClassDef node and inserts a docstring if missing."""
        # Check if the class already has a docstring
        if (node.body and
                isinstance(node.body[0], ast.Expr) and
                isinstance(node.body[0].value, (ast.Constant, ast.Str))):
            # Docstring already exists, do nothing
            return node

        # Generate a new docstring
        new_docstring_content = _generate_class_docstring(node)

        # Create an AST node for the docstring
        docstring_expr = ast.Expr(value=ast.Constant(value=new_docstring_content))

        # Copy line and column information from the class node
        ast.copy_location(docstring_expr, node)

        # Insert the docstring at the beginning of the class body
        node.body.insert(0, docstring_expr)

        ast.fix_missing_locations(node)

        return node

print("Defined DocstringInjector class for inserting docstrings into AST nodes.")
```

```python
import ast

def _get_type_string(annotation_node):
    """Helper to extract a readable type string from an AST annotation node."""
    if annotation_node is None:
        return 'Any' # Default if no annotation is present
    if isinstance(annotation_node, ast.Name):
        return annotation_node.id
    if isinstance(annotation_node, (ast.Constant, ast.Str)): # ast.Str for older Python versions
        return str(annotation_node.value)
    if isinstance(annotation_node, ast.Subscript):
        type_name = _get_type_string(annotation_node.value)
        if isinstance(annotation_node.slice, ast.Index):
            inner_type = _get_type_string(annotation_node.slice.value)
            return f"{type_name}[{inner_type}]"
        elif isinstance(annotation_node.slice, ast.Tuple):
            inner_types = [_get_type_string(elt) for elt in annotation_node.slice.elts]
            return f"{type_name}[{', '.join(inner_types)}]"
    return 'Any' # Fallback for complex or unhandled types

def _generate_function_docstring(func_node: ast.FunctionDef) -> str:
    """Generates a Google-style docstring content for a function node (without triple quotes)."""
    func_name = func_node.name
    content = []

    content.append(f"A brief description of the `{func_name}` function.")

    # Parameters
    params_exist = bool(func_node.args.posonlyargs or func_node.args.args or func_node.args.kwonlyargs)
    if params_exist:
        content.append("") # Blank line before Args
        content.append("Args:")
        for arg in func_node.args.posonlyargs + func_node.args.args + func_node.args.kwonlyargs:
            arg_name = arg.arg
            arg_type = _get_type_string(arg.annotation)
            content.append(f"    {arg_name} ({arg_type}): Description of `{arg_name}`.")

    # Return type
    return_type = _get_type_string(func_node.returns)
    content.append("") # Blank line before Returns
    content.append("Returns:")
    content.append(f"    {return_type}: Description of the return value.")

    # Examples
    content.append("") # Blank line before Examples
    content.append("Examples:")
    content.append(f"    >>> # Example usage for `{func_name}`")

    return '\n'.join(content)

def _generate_class_docstring(class_node: ast.ClassDef) -> str:
    """Generates a Google-style docstring content for a class node (without triple quotes)."""
    class_name = class_node.name
    docstring_content = f"A class representing a {class_name}.\n\n    This class is designed to [briefly describe its purpose]."
    return docstring_content

print("Redefined helper functions '_generate_function_docstring' and '_generate_class_docstring' to return raw docstring content.")
```

```python
import ast

class DocstringInjector(ast.NodeTransformer):
    """Inserts Google-style docstrings into function and class definitions if they don't already have one."""

    def visit_FunctionDef(self, node: ast.FunctionDef) -> ast.FunctionDef:
        """Visits a FunctionDef node and inserts a docstring if missing."""
        # Check if the function already has a docstring using the helper function
        if not _has_docstring(node.body):
            # Generate a new docstring (raw content without triple quotes)
            new_docstring_content = _generate_function_docstring(node)

            # Create an AST node for the docstring
            docstring_expr = ast.Expr(value=ast.Constant(value=new_docstring_content))

            # Copy line and column information from the function node for proper placement
            ast.copy_location(docstring_expr, node)

            # Insert the docstring at the beginning of the function body
            node.body.insert(0, docstring_expr)

            ast.fix_missing_locations(node)

        # Ensure to visit nested functions/classes within this function
        self.generic_visit(node)
        return node

    def visit_ClassDef(self, node: ast.ClassDef) -> ast.ClassDef:
        """Visits a ClassDef node and inserts a docstring if missing."""
        # Check if the class already has a docstring using the helper function
        if not _has_docstring(node.body):
            # Generate a new docstring (raw content without triple quotes)
            new_docstring_content = _generate_class_docstring(node)

            # Create an AST node for the docstring
            docstring_expr = ast.Expr(value=ast.Constant(value=new_docstring_content))

            # Copy line and column information from the class node
            ast.copy_location(docstring_expr, node)

            # Insert the docstring at the beginning of the class body
            node.body.insert(0, docstring_expr)

            ast.fix_missing_locations(node)

        # Ensure to visit nested functions/classes within this class (e.g., methods)
        self.generic_visit(node)
        return node

print("Updated DocstringInjector class to handle raw docstring content and recursive visits.")
```
