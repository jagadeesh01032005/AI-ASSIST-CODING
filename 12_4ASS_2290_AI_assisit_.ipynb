```python
def bubble_sort(scores):
    n = len(scores)
    # Outer loop for passes through the list
    for i in range(n - 1):
        swapped = False  # Flag to detect if any swaps occurred in this pass
        # Inner loop for comparisons and swaps in the current pass
        for j in range(0, n - i - 1):
            # Compare adjacent elements
            if scores[j] > scores[j + 1]:
                # If current element is greater than the next, swap them
                scores[j], scores[j + 1] = scores[j + 1], scores[j]
                swapped = True  # Mark that a swap occurred
        # Early termination condition: if no two elements were swapped
        # by inner loop, then the list is sorted
        if not swapped:
            break
    return scores

# Sample input of student scores
student_scores = [85, 72, 93, 65, 78, 90, 88, 70, 95]

print("Original scores:", student_scores)

# Sort the scores using Bubble Sort
sorted_scores = bubble_sort(list(student_scores)) # Create a copy to keep original untouched

print("Sorted scores:", sorted_scores)
```

```python
def insertion_sort(scores):
    # Traverse through 1 to len(scores)
    for i in range(1, len(scores)):
        key = scores[i] # Current element to be compared
        j = i - 1 # Index of the previous element

        # Move elements of scores[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < scores[j]:
            scores[j + 1] = scores[j]
            j -= 1
        scores[j + 1] = key # Place key at its correct position
    return scores

print("Defined the insertion_sort function.")
```

```python
def linear_search(roll_numbers, target_roll_number):
    """
    Searches for a target roll number in a list using linear search.

    Args:
        roll_numbers (list): A list of student roll numbers.
        target_roll_number (int): The roll number to search for.

    Returns:
        int: The index of the target roll number if found, otherwise -1.
    """
    for index, roll_number in enumerate(roll_numbers):
        if roll_number == target_roll_number:
            return index
    return -1

print("Defined the linear_search function.")
```

```python
def binary_search(roll_numbers, target_roll_number):
    """
    Searches for a target roll number in a *sorted* list using binary search.

    Args:
        roll_numbers (list): A sorted list of student roll numbers.
        target_roll_number (int): The roll number to search for.

    Returns:
        int: The index of the target roll number if found, otherwise -1.
    """
    low = 0
    high = len(roll_numbers) - 1

    while low <= high:
        mid = (low + high) // 2

        if roll_numbers[mid] == target_roll_number:
            return mid
        elif roll_numbers[mid] < target_roll_number:
            low = mid + 1
        else:
            high = mid - 1
    return -1

print("Defined the binary_search function.")
```

```python
def quick_sort(arr):
    """
    Sorts a list of elements using the Quick Sort algorithm.

    Quick Sort is a highly efficient, comparison-based sorting algorithm that
    uses a divide-and-conquer approach. It works by selecting a 'pivot' element
    from the array and partitioning the other elements into two sub-arrays,
    according to whether they are less than or greater than the pivot.
    The sub-arrays are then recursively sorted.

    Args:
        arr (list): The list of elements to be sorted. It can contain any
                    comparable data types (e.g., numbers).

    Returns:
        list: A new list containing the same elements as the input list,
              but sorted in ascending order.

    How Recursion Works:
    1. Base Case: If the list has 0 or 1 element, it is already sorted,
       so it is returned directly. This stops the recursion.
    2. Divide: A pivot element is chosen (here, the last element). The rest
       of the list is then divided into two sub-lists: `less` (elements
       smaller than the pivot) and `greater` (elements greater than or equal
       to the pivot).
    3. Conquer: The `quick_sort` function is recursively called on both the
       `less` and `greater` sub-lists. This process continues until the
       base case is met for all sub-lists.
    4. Combine: Once the recursive calls return the sorted `less` and `greater`
       sub-lists, they are concatenated together with the pivot in between
       ([sorted_less] + [pivot] + [sorted_greater]) to produce the final
       sorted list.
    """
    # 2. Implement the base case for the recursion
    if len(arr) <= 1:
        return arr

    # 3. Choose a pivot element (last element) and remove it from the list for partitioning
    pivot = arr.pop() # Removes the last element and returns it

    # 4. Create two empty lists for partitioning
    less = []
    greater = []

    # 5. Iterate and append to less or greater based on comparison with pivot
    for element in arr:
        if element < pivot:
            less.append(element)
        else:
            greater.append(element)

    # 6. Recursively call quick_sort on the less and greater lists
    # 7. Combine the sorted less list, the pivot, and the sorted greater list
    return quick_sort(less) + [pivot] + quick_sort(greater)

print("Defined the quick_sort function.")
```

```python
def merge_sort(arr):
    """
    Sorts a list of elements using the Merge Sort algorithm.

    Merge Sort is an efficient, comparison-based, divide-and-conquer sorting algorithm.
    It works by recursively dividing an unsorted list into n sublists, each containing
    one element (a list of one element is considered sorted). Then, it repeatedly merges
    sublists to produce new sorted sublists until there is only one sorted list remaining.

    Args:
        arr (list): The list of elements to be sorted. It can contain any
                    comparable data types (e.g., numbers).

    Returns:
        list: A new list containing the same elements as the input list,
              but sorted in ascending order.

    How Recursion Works:
    1. Base Case: If the list has 0 or 1 element, it is considered sorted and returned
       directly. This is the termination condition for the recursion.
    2. Divide: If the list has more than one element, it is divided into two roughly
       equal halves (left_half and right_half) at the middle index.
    3. Conquer: The `merge_sort` function is recursively called on both `left_half`
       and `right_half`. This continues to divide the lists until the base case is reached
       for all sub-lists.
    4. Combine: Once the recursive calls return the sorted sub-lists, a helper function
       `_merge` is used to combine these two sorted halves back into a single, fully sorted list.
    """
    # 2. Implement the base case for the recursion
    if len(arr) <= 1:
        return arr

    # 3. If the list has more than one element, find the middle index
    mid = len(arr) // 2

    # 4. Create two sub-lists, left_half and right_half
    left_half = arr[:mid]
    right_half = arr[mid:]

    # 5. Recursively call merge_sort on left_half and right_half
    sorted_left = merge_sort(left_half)
    sorted_right = merge_sort(right_half)

    # 6. Implement a helper function `_merge` and call it to combine results
    # This helper function is typically nested or defined separately.
    # For clarity, let's define it as a local helper or assume it's external for now,
    # but include its logic here.
    def _merge(left, right):
        merged_list = []
        i = j = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                merged_list.append(left[i])
                i += 1
            else:
                merged_list.append(right[j])
                j += 1

        # Append any remaining elements from left or right (if one list is longer)
        merged_list.extend(left[i:])
        merged_list.extend(right[j:])
        return merged_list

    return _merge(sorted_left, sorted_right)

print("Defined the merge_sort function.")
```

```python
def find_duplicates_naive(user_ids):
    """
    Detects duplicate user IDs in a list using a naive approach with nested loops.

    This function iterates through the list of user IDs, comparing each element
    with every subsequent element to find duplicates. It collects and returns
    unique duplicate IDs found.

    Args:
        user_ids (list): A list of user IDs to check for duplicates.

    Returns:
        list: A list of unique duplicate user IDs found in the input list.
    """
    duplicates = []
    n = len(user_ids)

    # Outer loop iterates from the first element up to, but not including, the last element
    for i in range(n - 1):
        # Inner loop iterates from the element *after* the current outer loop element
        # up to the end of the user_ids list
        for j in range(i + 1, n):
            # Compare the element from the outer loop with the element from the inner loop
            if user_ids[i] == user_ids[j]:
                # If the two elements are equal and the element is not already in the duplicates list,
                # add it to the duplicates list.
                if user_ids[i] not in duplicates:
                    duplicates.append(user_ids[i])

    return duplicates

print("Defined the 'find_duplicates_naive' function.")

# Example usage for testing the function
# sample_user_ids = [101, 102, 103, 101, 104, 105, 102, 106, 101]
# found_duplicates = find_duplicates_naive(sample_user_ids)
# print(f"Sample User IDs: {sample_user_ids}")
# print(f"Found duplicates: {found_duplicates}") # Expected: [101, 102]
```

```python
def find_duplicates_optimized(user_ids):
    """
    Detects duplicate user IDs in a list using an optimized approach with a set.

    This function efficiently identifies duplicate user IDs by using a set to track
    seen IDs, providing O(1) average time complexity for lookups. It collects
    and returns unique duplicate IDs found.

    Args:
        user_ids (list): A list of user IDs to check for duplicates.

    Returns:
        list: A list of unique duplicate user IDs found in the input list.
    """
    seen_ids = set() # To store unique IDs encountered
    duplicates = []  # To store unique duplicate IDs found

    for user_id in user_ids:
        # If the user_id has been seen before, it's a duplicate
        if user_id in seen_ids:
            # Add to duplicates list only if not already added (to ensure unique duplicates)
            if user_id not in duplicates:
                duplicates.append(user_id)
        else:
            # If not seen, add to the set of seen IDs
            seen_ids.add(user_id)

    return duplicates

print("Defined the 'find_duplicates_optimized' function.")

# Example usage for testing the function
# sample_user_ids = [101, 102, 103, 101, 104, 105, 102, 106, 101]
# found_duplicates = find_duplicates_optimized(sample_user_ids)
# print(f"Sample User IDs: {sample_user_ids}")
# print(f"Found duplicates (optimized): {found_duplicates}") # Expected: [101, 102]
```

```python
import random

# 1. Define a list of possible user ID ranges or starting points
# To ensure duplicates are likely, we'll keep the range relatively small
# and generate more IDs than the range size, or explicitly add duplicates.

def generate_user_ids_with_duplicates(num_ids=20, id_range=(1000, 1020), num_duplicates=5):
    user_ids = []
    # Generate initial unique IDs
    unique_ids_pool = list(range(id_range[0], id_range[1] + 1))

    # Ensure we don't try to generate more unique IDs than available in the pool
    actual_unique_ids_to_generate = min(num_ids - num_duplicates, len(unique_ids_pool))

    # Add a base set of unique IDs
    user_ids.extend(random.sample(unique_ids_pool, actual_unique_ids_to_generate))

    # Add some additional IDs, which will likely introduce duplicates
    # or ensure some duplicates are present by picking from the existing set
    remaining_ids_to_generate = num_ids - len(user_ids)
    for _ in range(remaining_ids_to_generate):
        # Randomly choose an ID from the original unique pool to ensure potential duplicates
        user_ids.append(random.choice(unique_ids_pool))

    random.shuffle(user_ids) # Shuffle to mix duplicates throughout the list
    return user_ids

# Generate a list of user IDs with duplicates
test_user_ids = generate_user_ids_with_duplicates(num_ids=25, id_range=(1000, 1015), num_duplicates=10)

print("Generated Test User IDs with Duplicates:", test_user_ids)
print(f"Total IDs: {len(test_user_ids)}")
print(f"Number of unique IDs if no duplicates were present (based on range): {1015 - 1000 + 1}")

# Verify some duplicates (optional, for debugging)
from collections import Counter
id_counts = Counter(test_user_ids)
duplicated_items = [id for id, count in id_counts.items() if count > 1]
print("Expected Duplicates (from generated list, for verification):", duplicated_items)
```
