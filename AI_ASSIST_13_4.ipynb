```python
def bubble_sort(scores):
    n = len(scores)
    # Outer loop for passes through the list
    for i in range(n - 1):
        swapped = False  # Flag to detect if any swaps occurred in this pass
        # Inner loop for comparisons and swaps within each pass
        for j in range(n - 1 - i):
            # Comparison: Check if the current element is greater than the next element
            if scores[j] > scores[j + 1]:
                # Swap: If elements are in the wrong order, swap them
                scores[j], scores[j + 1] = scores[j + 1], scores[j]
                swapped = True  # Mark that a swap occurred
        # Early Termination Condition: If no two elements were swapped by the inner loop,
        # then the list is sorted, and we can stop early.
        if not swapped:
            break
    return scores
```

```python
# Sample student scores
student_scores = [85, 72, 90, 65, 78, 92, 70, 88]

print("Original scores:", student_scores)

# Sort the scores using the bubble_sort function
sorted_scores = bubble_sort(student_scores.copy()) # Use .copy() to keep original list intact for display

print("Sorted scores:", sorted_scores)

# Example with an already sorted list (best case)
already_sorted_scores = [60, 70, 80, 90, 100]
print("\nOriginal already sorted scores:", already_sorted_scores)
print("Sorted already sorted scores:", bubble_sort(already_sorted_scores.copy()))

# Example with a reverse sorted list (worst case)
reverse_sorted_scores = [100, 90, 80, 70, 60]
print("\nOriginal reverse sorted scores:", reverse_sorted_scores)
print("Sorted reverse sorted scores:", bubble_sort(reverse_sorted_scores.copy()))
```

```python
def insertion_sort(scores):
    n = len(scores)
    # Traverse through 1 to n (starting from the second element)
    for i in range(1, n):
        key = scores[i]  # Current element to be inserted
        j = i - 1

        # Move elements of scores[0..i-1], that are greater than key,
        # to one position ahead of their current position
        while j >= 0 and key < scores[j]:
            scores[j + 1] = scores[j]
            j -= 1
        scores[j + 1] = key  # Place key at its correct position
    return scores
```

```python
import random

def generate_nearly_sorted_roll_numbers(size=100, num_swaps=5):
    # 1. Generate a perfectly sorted list of student roll numbers
    roll_numbers = list(range(1, size + 1))

    # 2. Introduce a small number of random swaps or disturbances
    for _ in range(num_swaps):
        idx1 = random.randint(0, size - 1)
        idx2 = random.randint(0, size - 1)
        # Ensure we are not swapping an element with itself
        while idx1 == idx2:
            idx2 = random.randint(0, size - 1)
        roll_numbers[idx1], roll_numbers[idx2] = roll_numbers[idx2], roll_numbers[idx1]

    return roll_numbers

# Store this nearly sorted list in a variable
nearly_sorted_roll_numbers = generate_nearly_sorted_roll_numbers(size=20, num_swaps=2)

# Print the generated nearly_sorted_roll_numbers list
print("Nearly Sorted Roll Numbers:", nearly_sorted_roll_numbers)
```

```python
def linear_search(scores, target_score):
    """
    Performs a linear search to find the index of a target score in a list of scores.

    Args:
        scores (list): A list of numerical scores to search through.
        target_score (int or float): The score to search for within the list.

    Returns:
        int: The index of the target_score if found, otherwise -1.
    """
    for i in range(len(scores)):
        if scores[i] == target_score:
            return i
    return -1
```

```python
def binary_search(scores, target_score):
    """
    Performs a binary search to find the index of a target score in a sorted list of scores.

    Args:
        scores (list): A sorted list of numerical scores to search through.
        target_score (int or float): The score to search for within the list.

    Returns:
        int: The index of the target_score if found, otherwise -1.
    """
    left, right = 0, len(scores) - 1

    while left <= right:
        mid = (left + right) // 2
        if scores[mid] == target_score:
            return mid
        elif scores[mid] < target_score:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
def quick_sort(arr, low=0, high=None):
    """
    Sorts a list of elements using the Quick Sort algorithm.

    Quick Sort is a highly efficient, comparison-based sorting algorithm.
    It works by selecting a 'pivot' element from the array and partitioning
    the other elements into two sub-arrays, according to whether they are
    less than or greater than the pivot. The sub-arrays are then sorted
    recursively.

    Args:
        arr (list): The list of numerical elements to be sorted in-place.
        low (int, optional): The starting index of the sub-array to be sorted.
                             Defaults to 0 (the beginning of the list).
        high (int, optional): The ending index of the sub-array to be sorted.
                              Defaults to len(arr) - 1 (the end of the list).

    Returns:
        list: The sorted list (although the sorting is done in-place).
    """
    if high is None:
        high = len(arr) - 1

    def partition(arr, low, high):
        # Choose the element at 'high' as the pivot
        pivot = arr[high]
        # Initialize a pointer 'i' to 'low - 1'
        i = low - 1

        # Iterate 'j' from 'low' to 'high - 1'
        for j in range(low, high):
            # If current element is smaller than or equal to pivot
            if arr[j] <= pivot:
                # Increment index of smaller element
                i += 1
                arr[i], arr[j] = arr[j], arr[i] # Swap elements

        # Swap arr[i + 1] with arr[high] (the pivot)
        arr[i + 1], arr[high] = arr[high], arr[i + 1]
        return i + 1  # Return the partitioning index

    # Base case for the recursion: if low is greater than or equal to high,
    # the sub-array is already sorted or has one element, so return.
    if low < high:
        # Call the partition helper function to get the pivot's final index
        pi = partition(arr, low, high)

        # Make two recursive calls to quick_sort
        # One for the sub-array to the left of the pivot
        quick_sort(arr, low, pi - 1)
        # One for the sub-array to the right of the pivot
        quick_sort(arr, pi + 1, high)
    return arr
```

```python
def merge_sort(arr):
    """
    Sorts a list of elements using the Merge Sort algorithm.

    Merge Sort is a divide-and-conquer algorithm that works by dividing
    an array into two halves, recursively sorting each half, and then
    merging the two sorted halves.

    Args:
        arr (list): The list of numerical elements to be sorted.

    Returns:
        list: A new list containing the sorted elements.
    """
    # Base case: if the list has 0 or 1 element, it is already sorted
    if len(arr) <= 1:
        return arr

    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    # Recursively sort both halves
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    # Merge the sorted halves
    return merge(left_half, right_half)

def merge(left, right):
    """
    Merges two sorted lists into a single sorted list.

    Args:
        left (list): The first sorted list.
        right (list): The second sorted list.

    Returns:
        list: A new list containing all elements from both input lists,
              sorted in ascending order.
    """
    merged_list = []
    i = j = 0

    # Compare elements from both lists and append the smaller one
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged_list.append(left[i])
            i += 1
        else:
            merged_list.append(right[j])
            j += 1

    # Append any remaining elements from the left list
    while i < len(left):
        merged_list.append(left[i])
        i += 1

    # Append any remaining elements from the right list
    while j < len(right):
        merged_list.append(right[j])
        j += 1

    return merged_list
```

```python
def find_duplicates_brute_force(data_list):
    """
    Detects duplicate elements in a list using a brute-force approach with nested loops.

    This function iterates through the list, comparing each element with every subsequent
    element to find identical values. If two identical elements are found at different
    indices, the element is considered a duplicate.

    Args:
        data_list (list): The list of items to search for duplicates.
                          Can contain any comparable data types.

    Returns:
        list: A list containing all duplicate elements found in the input list.
              If an element appears more than twice, it will be added multiple times
              to the returned list (e.g., [1, 2, 2, 2] -> [2, 2]). The order of
              duplicates in the returned list corresponds to their first detection.
    """
    duplicates = []
    n = len(data_list)

    # Outer loop iterates from the first element to the second-to-last element
    for i in range(n - 1):
        # Inner loop iterates from the element after the outer loop's current element
        # to the last element
        for j in range(i + 1, n):
            # Compare elements
            if data_list[i] == data_list[j]:
                # If identical elements are found at different indices, add to duplicates
                duplicates.append(data_list[i])
                # Optional: If you only want each unique duplicate to appear once
                # if data_list[i] not in duplicates:
                #     duplicates.append(data_list[i])
    return duplicates
```

```python
print("\n--- Demonstrating Brute-Force Duplicate Detection ---")

# Sample data with duplicates
list1 = [1, 2, 3, 4, 2, 5, 1, 6, 7, 3]
print(f"Original List 1: {list1}")
duplicates1 = find_duplicates_brute_force(list1)
print(f"Duplicates in List 1: {duplicates1}")

# Sample data with no duplicates
list2 = [10, 20, 30, 40, 50]
print(f"\nOriginal List 2: {list2}")
duplicates2 = find_duplicates_brute_force(list2)
print(f"Duplicates in List 2: {duplicates2}")

# Sample data with multiple occurrences of the same duplicate
list3 = ['a', 'b', 'c', 'a', 'd', 'b', 'a']
print(f"\nOriginal List 3: {list3}")
duplicates3 = find_duplicates_brute_force(list3)
print(f"Duplicates in List 3: {duplicates3}")
```

```python
def find_duplicates_optimized(data_list):
    """
    Detects duplicate elements in a list using an optimized approach with a set.

    This function iterates through the list once. It uses a `seen` set to keep track
    of elements encountered so far and a `duplicates` set to store elements that
    have been identified as duplicates. This approach provides better performance
    compared to the brute-force method for larger datasets.

    Args:
        data_list (list): The list of items to search for duplicates.
                          Can contain any hashable data types.

    Returns:
        list: A list containing unique duplicate elements found in the input list.
              The order of elements in the returned list is not guaranteed.
    """
    seen = set()
    duplicates = set()

    for item in data_list:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)

    return list(duplicates)
```
