```python
def login(user, password, credentials):
    """Authenticates a user against a dictionary of credentials.

    Args:
        user (str): The username provided by the user.
        password (str): The password provided by the user.
        credentials (dict): A dictionary where keys are usernames (str) and values are passwords (str).

    Returns:
        bool: True if the user's password matches the stored credentials, False otherwise.
    """
    return credentials.get(user) == password

# Example usage:
user_credentials = {"admin": "secret123", "guest": "pass"}
print(f"Admin login attempt: {login('admin', 'secret123', user_credentials)}")
print(f"Guest login attempt with wrong password: {login('guest', 'wrongpass', user_credentials)}")
```

```python
def login(user, password, credentials):
    # This function attempts to log in a user.
    # It compares the provided 'user' and 'password' against 'credentials'.
    # 'user' is a string representing the username.
    # 'password' is a string representing the user's password.
    # 'credentials' is a dictionary: {'username': 'password'}.
    # It returns True if the provided password matches the stored password for the given user,
    # otherwise, it returns False.
    stored_password = credentials.get(user) # Retrieve the stored password for the given user.
    return stored_password == password # Compare the provided password with the stored one.

# Example usage:
user_credentials = {"dev": "devpass", "test": "testpass"}
print(f"Dev login attempt: {login('dev', 'devpass', user_credentials)}")
```

```python
def login(user, password, credentials):
    """Authenticates a user against a dictionary of credentials.

    Args:
        user (str): The username provided by the user attempting to log in.
        password (str): The password provided by the user.
        credentials (dict): A dictionary containing valid username-password pairs.
                            Keys are usernames (str), and values are their corresponding passwords (str).

    Returns:
        bool: True if the `user` exists in `credentials` and the provided `password` matches
              the stored password for that user; False otherwise.

    Example:
        >>> user_db = {"alice": "securepass", "bob": "anotherpass"}
        >>> login("alice", "securepass", user_db)
        True
        >>> login("bob", "wrongpass", user_db)
        False
        >>> login("charlie", "anypass", user_db) # User 'charlie' does not exist
        False
    """
    return credentials.get(user) == password

# Example usage:
user_credentials = {"manager": "mypass", "analyst": "anapass"}
print(f"Manager login attempt: {login('manager', 'mypass', user_credentials)}")
```

```python
%%writefile calculator.py
def add(a, b):
    """Adds two numbers.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The sum of a and b.
    """
    return a + b

def subtract(a, b):
    """Subtracts the second number from the first.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The difference between a and b (a - b).
    """
    return a - b

def multiply(a, b):
    """Multiplies two numbers.

    Args:
        a (int or float): The first number.
        b (int or float): The second number.

    Returns:
        int or float: The product of a and b.
    """
    return a * b

def divide(a, b):
    """Divides the first number by the second.

    Args:
        a (int or float): The numerator.
        b (int or float): The denominator.

    Returns:
        int or float: The quotient of a and b (a / b).

    Raises:
        ValueError: If the denominator b is zero.
    """
    if b == 0:
        raise ValueError("Cannot divide by zero.")
    return a / b

print("calculator.py created successfully with add, subtract, multiply, and divide functions.")
```

```python
%%writefile conversion.py
def decimal_to_binary(n):
    """Converts a decimal integer to its binary string representation.

    Args:
        n (int): The decimal integer to convert.

    Returns:
        str: The binary representation of the decimal number as a string.

    Example:
        >>> decimal_to_binary(10)
        '1010'
    """
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n < 0:
        return '-' + bin(abs(n))[2:]
    return bin(n)[2:]

def binary_to_decimal(b):
    """Converts a binary string to its decimal integer representation.

    Args:
        b (str): The binary string to convert. Can optionally start with '-'.

    Returns:
        int: The decimal integer representation of the binary string.

    Raises:
        ValueError: If the input string is not a valid binary number.

    Example:
        >>> binary_to_decimal('1010')
        10
        >>> binary_to_decimal('-101')
        -5
    """
    if not isinstance(b, str):
        raise TypeError("Input must be a string.")
    if not all(char in '01' for char in b.lstrip('-')):
        raise ValueError("Input string must be a valid binary number.")

    is_negative = False
    if b.startswith('-'):
        is_negative = True
        b = b[1:]

    decimal_val = int(b, 2)
    return -decimal_val if is_negative else decimal_val

def decimal_to_hexadecimal(n):
    """Converts a decimal integer to its hexadecimal string representation.

    Args:
        n (int): The decimal integer to convert.

    Returns:
        str: The hexadecimal representation of the decimal number as a string, prefixed with '0x'.

    Example:
        >>> decimal_to_hexadecimal(255)
        '0xff'
    """
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    return hex(n)

print("conversion.py created successfully with decimal_to_binary, binary_to_decimal, and decimal_to_hexadecimal functions.")
```
