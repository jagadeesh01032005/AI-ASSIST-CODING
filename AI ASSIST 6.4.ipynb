```python
class Student:
    def __init__(self, name, roll_number, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    # Copilot: method to display student details
    def display_details(self):
        print(f"Name: {self.name}")
        print(f"Roll Number: {self.roll_number}")
        print(f"Marks: {self.marks}")

    # Copilot: method to check if student's marks are above class average
    def check_performance(self, class_average):
        if self.marks > class_average:
            return f"{self.name} with Roll Number {self.roll_number} has marks ({self.marks}) above the class average ({class_average})."
        elif self.marks < class_average:
            return f"{self.name} with Roll Number {self.roll_number} has marks ({self.marks}) below the class average ({class_average})."
        else:
            return f"{self.name} with Roll Number {self.roll_number} has marks ({self.marks}) equal to the class average ({class_average})."
```

**SUMMARY**

Task 01 involved creating a Python class named Student with attributes for name, roll_number, and marks. Key methods included display_details() to show student information and check_performance(class_average) to compare a student's marks against a class average. This method used conditional logic to return an appropriate message indicating if the student's performance was above, below, or equal to the average. Sample usage demonstrated the creation of student objects and the functionality of these methods, including output for displaying details and performance status.

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self.balance = initial_balance

    # Copilot: method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposit of ${amount:.2f} successful. New balance: ${self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    # Copilot: method to withdraw money, preventing withdrawals when balance is insufficient
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be positive.")
        elif self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawal of ${amount:.2f} successful. New balance: ${self.balance:.2f}")
        else:
            print(f"Insufficient funds. Current balance: ${self.balance:.2f}. Attempted withdrawal: ${amount:.2f}")
```

```python
# Create a bank account
my_account = BankAccount("Alice Wonderland", 1000)
print(f"Initial balance for {my_account.account_holder}: ${my_account.balance:.2f}")

print("\n--- Transactions ---")

# Deposit money
my_account.deposit(500)

# Withdraw money successfully
my_account.withdraw(200)

# Attempt to withdraw more than available balance
my_account.withdraw(1500)

# Attempt to deposit invalid amount
my_account.deposit(-50)

# Attempt to withdraw invalid amount
my_account.withdraw(0)

print(f"\nFinal balance for {my_account.account_holder}: ${my_account.balance:.2f}")
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = [] # An empty list to store items

    # Copilot: method to add items to the cart
    def add_item(self, name, price, quantity=1):
        item = {"name": name, "price": price, "quantity": quantity}
        self.items.append(item)
        print(f"Added {quantity} x {name} to the cart.")

    # Copilot: method to remove items from the cart
    def remove_item(self, name):
        initial_len = len(self.items)
        self.items = [item for item in self.items if item["name"] != name]
        if len(self.items) < initial_len:
            print(f"Removed all instances of {name} from the cart.")
        else:
            print(f"{name} not found in the cart.")

    # Copilot: method to calculate the total bill using a loop
    def calculate_total(self):
        total = 0
        for item in self.items:
            total += item["price"] * item["quantity"]
        return total

    # Copilot: method to apply conditional discounts (e.g., discount if total exceeds a certain amount)
    def apply_discount(self, discount_threshold, discount_percentage):
        total = self.calculate_total()
        if total > discount_threshold:
            discount_amount = total * (discount_percentage / 100)
            discounted_total = total - discount_amount
            print(f"Total ({total:.2f}) exceeds discount threshold (${discount_threshold:.2f}). Applied {discount_percentage}% discount.")
            return discounted_total
        else:
            print(f"Total ({total:.2f}) does not meet discount threshold (${discount_threshold:.2f}). No discount applied.")
            return total
```

```python
# Create a shopping cart
my_cart = ShoppingCart()

print("--- Adding items ---")
my_cart.add_item("Laptop", 1200, 1)
my_cart.add_item("Mouse", 25, 2)
my_cart.add_item("Keyboard", 75)

print("\n--- Current Cart Items ---")
for item in my_cart.items:
    print(f"- {item['name']} (Price: ${item['price']:.2f}, Quantity: {item['quantity']})")

print("\n--- Calculating Total ---")
initial_total = my_cart.calculate_total()
print(f"Initial Total Bill: ${initial_total:.2f}")

print("\n--- Applying Discount ---")
discount_threshold = 1000
discount_percentage = 10
final_total_after_discount = my_cart.apply_discount(discount_threshold, discount_percentage)
print(f"Final Total after Discount: ${final_total_after_discount:.2f}")

print("\n--- Removing an Item ---")
my_cart.remove_item("Mouse")

print("\n--- Cart after Removal ---")
if not my_cart.items:
    print("Cart is empty.")
else:
    for item in my_cart.items:
        print(f"- {item['name']} (Price: ${item['price']:.2f}, Quantity: {item['quantity']})")

print("\n--- Recalculating Total after Removal ---")
updated_total = my_cart.calculate_total()
print(f"Updated Total Bill: ${updated_total:.2f}")

print("\n--- Re-applying Discount with new total ---")
final_total_after_reapply = my_cart.apply_discount(discount_threshold, discount_percentage)
print(f"Final Total after Re-apply Discount: ${final_total_after_reapply:.2f}")
```

**SUMMARY**

Task 05 focused on designing a ShoppingCart class for an e-commerce system. This class included methods to add_item (with optional quantity), remove_item (handling all instances of a named item), calculate_total using a loop, and apply_discount conditionally based on a total threshold. The implementation utilized object-oriented programming, list manipulations, loops, and conditional logic to manage cart functionality and calculate the final bill, including discount application. Sample usage demonstrated adding, removing, calculating totals, and applying discounts.
